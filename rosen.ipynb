```python
import numpy as np
from scipy.optimize import minimize

# Importing the Rosenbrock function, wich computed as f(x) = sum(100.0*(x[1:] - x[:-1]**2.0)**2.0 + (1 - x[:-1])**2.0
from scipy.optimize import rosen
```

### The Rosenbrock function
\- is a non-convex function, also known as _Rosenbrock's valley_ or _Rosenbrock's banana_ function.
<br>The global minimum is inside a long, narrow, parabolic shaped flat valley. 

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
%matplotlib inline

fig = plt.figure(figsize=[15, 10])
ax = fig.gca(projection='3d')
ax.view_init(30, 30)

X = np.arange(-2, 2, 0.1)
Y = np.arange(-2, 4, 0.1)
X, Y = np.meshgrid(X, Y)
Z = rosen(np.array([X,Y]))

surf = ax.plot_surface(X, Y, Z, cmap='magma')
```

### BFGS
The BroydenFletcherGoldfarbShanno (BFGS) algorithm is an iterative method for solving unconstrained nonlinear optimization problems.

```python
x0 = np.array([1.3, 0.7, 0.8, 1.9, 1.2])
# If method is not given, chosen to be one of BFGS, L-BFGS-B, SLSQP, depending if the problem has constraints or bounds
res = minimize(rosen, x0, options={'disp': True})
print(res.x)
```

### The NelderMead method
The NelderMead method (also _downhill simplex method, amoeba method, or polytope method_) is a commonly applied numerical method used to find the minimum or maximum of an objective function in a multidimensional space.
<br><br>It is a direct search method (based on function comparison) and is often applied to nonlinear optimization problems for which derivatives may not be known. However, the NelderMead technique is a heuristic search method that can converge to non-stationary points on problems that can be solved by alternative methods.

```python
# Nelder-Mead method
x0 = np.array([1.3, 0.7, 0.8, 1.9, 1.2])
res = minimize(rosen, x0, method='nelder-mead',
    options={'xtol': 1e-8, 'disp': True})
print(res.x)
```
